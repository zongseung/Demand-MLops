```python
from pathlib import Path
import pandas as pd

# 폴더 경로
base_dir = Path("/mnt/nvme/open-stef/collect_demand")

# 파일 패턴에 맞는 CSV 모두 탐색
csv_files = sorted(base_dir.glob("asos_*.csv"))

print(f"불러올 파일 수: {len(csv_files)}")
for f in csv_files:
    print(f)

# 한꺼번에 읽고 row 기준으로 이어붙이기
df_list = [pd.read_csv(f, encoding="utf-8-sig") for f in csv_files]
asos_all = pd.concat(df_list, ignore_index=True)

print(f"\n최종 데이터 shape: {asos_all.shape}")

# 저장
asos_all.to_csv(base_dir / "asos_all_merged.csv", index=False, encoding="utf-8-sig")
```
